```python
x= df.drop(columns = ['price'], axis = 1)
y = df['price']
```

```python
x
```

```python
df.columns
```

```python
x_encoded = pd.get_dummies(x,columns=['model' , 'transmission' , 'fuelType'], drop_first=True)
```

```python
x_encoded
```

```python
x_encoded = x_encoded.astype(int)
```

```python
# Keep dtypes as produced by get_dummies; avoid forced integer casting
x_encoded = x_encoded.copy()
```

```python
x_encoded
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

xlable = x.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    xlable[col] = le.fit_transform(xlable[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
xlable
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
```

```python
x_encoded[numeric_cols] = scaler.fit_transform(x_encoded[numeric_cols])
```

```python
x_encoded
```

```python
xlable.columns
```

```python
xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
xlable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = x_encoded.shape[0]  # number of observations
p = x_encoded.shape[1]  # number of predictors
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
adjusted_r2
```
